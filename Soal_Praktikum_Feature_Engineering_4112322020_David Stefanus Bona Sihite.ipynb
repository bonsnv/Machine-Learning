  # Nomor 3
  a. Mengubah Fitur kategorikal (fuel_type, transmission, car_brand, color) menjadi numerik menggunakan **One-Hot Encoding**

```python
# Melakukan One-Hot Encoding pada fitur kategorikal
df_encoded = pd.get_dummies(df, columns=["fuel_type", "transmission", "car_brand", "color"], drop_first=True)

# Menampilkan hasil setelah encoding
df_encoded.head()
```

b. Dampak One-Hot Encoding terhadap Jumlah Fitur:
- Jumlah fitur bertambah signifikan karena setiap kategori unik dalam fitur awal dibuat menjadi kolom baru.
- Dimensi dataset meningkat
- Model machine learning bisa menjadi lebih kompleks

c. One-Hot Encoding tidak efektif jika suatu fitur memiliki 100 kategori unik karena:
- Meningkatkan Dimensi Data Secara Drastis → Jika ada 100 kategori, maka dataset akan memiliki tambahan 99 kolom baru.
- Membutuhkan Lebih Banyak Memori dan Waktu Komputasi → Model bisa menjadi lambat dan sulit diinterpretasikan.
- Menghasilkan Banyak Kolom Sparsity (Banyak Nol) → Bisa menyebabkan masalah dalam model berbasis regresi atau jaringan saraf.

Alternatif One-Hot Encoding untuk banyak kategori:
- Target Encoding(Mean Encoding)
- Ordinal Encoding
- Embedding Encoding
- Frequency Encoding

  # Nomor 4
a. Scaling pada fitur numerik (milage, engine_size, num_owners) menggunakan **StandardScaler**

```python
from sklearn.preprocessing import StandardScaler

# Fitur numerik yang akan di-scaling
numeric_features = ["mileage", "engine_size", "num_owners"]

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Melakukan scaling
df_encoded[numeric_features] = scaler.fit_transform(df_encoded[numeric_features])

# Menampilkan hasil setelah scaling
df_encoded.head()
```

b. Perbedaan StandardScaler, MinMaxScaler, dan RobustScaler:
1. **StandardScaler:**
- Mengonversi data agar memiliki mean = 0 dan standar deviasi = 1.
- Cocok untuk data yang terdistribusi normal.
- Tidak cocok jika ada outlier, karena outlier akan tetap berpengaruh.
2. **MinMaxScaler:**
- Mengubah data ke skala 0 hingga 1 berdasarkan nilai minimum dan maksimum.
- Cocok untuk data yang tidak memiliki outlier, karena rentang nilai sangat dipengaruhi oleh titik ekstrem.
- Digunakan pada model yang sensitif terhadap skala data, seperti Neural Networks atau K-Nearest Neighbors (KNN).
3. **RobustScaler:**
- Memanfaatkan median (Q2) dan interquartile range (IQR) untuk scaling.
- Cocok untuk data dengan banyak outlier, karena tidak terpengaruh oleh nilai ekstrem.

c. Dalam Kondisi apa menggunakan MinMaxScaler daripada StandardScaler?
- Jika distribusi data tidak normal dan memiliki rentang tetap.
- Jika menggunakan model berbasis jarak seperti KNN atau SVM, karena mereka lebih sensitif terhadap skala data.
- Jika menggunakan Neural Networks, karena rentang nilai 0-1 bisa mempercepat konvergensi selama training.

a. Buat Fitur *milage_per_year* dengan membagi *milage* dengan *car_age*.

```python
# Menambahkan fitur mileage_per_year
df_encoded["mileage_per_year"] = df["mileage"] / df_encoded["car_age"]

# Menampilkan hasil dengan fitur baru
df_encoded[["mileage", "car_age", "mileage_per_year"]].head()
```

b. Mengapa mileage_per_year lebih informatif dibandingkan mileage saja?
- Menormalkan Pengaruh Usia Mobil
- Lebih Representatif untuk Analisis Harga dan Kelayakan Mobil
- Mempermudah Model Prediksi

Kesimpulannya, *mileage_per_year* lebih informatif karena mencerminkan intensitas penggunaan mobil, bukan hanya total jarak tempuhnya.
